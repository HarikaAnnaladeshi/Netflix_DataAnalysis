Categorizing Vote_Average column
We would ut the Vote_Average values and make 4 categories: popular, average, below average, not_popular to describe it more using categorize_col() function provided above

```python
def categorize_col(df, col,labels):
    edges=[df[col].describe()['min'],
            df[col].describe()['25%'],
            df[col].describe()['50%'],
            df[col].describe()['75%'],
            df[col].describe()['max']]
    df[col]=pd.cut(df[col],edges,labels=labels,duplicates='drop')
    return df
```

```python
labels=['not_popular','below_average','average','popular']
categorize_col(df,'Vote_Average',labels)
```

We'd split genres into a list amd then explode our dataframe to have only one genre per row for ezch movie

```python
# Split the string by comma and remove surrounding whitespace for each genre
df['Genre'] = df['Genre'].str.split(',')
df['Genre'] = df['Genre'].apply(lambda x: [g.strip() for g in x])

# Now explode the list into separate rows
df = df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
#casting column into category
df['Genre']=df['Genre'].astype('category')
df['Genre'].dtype
```
